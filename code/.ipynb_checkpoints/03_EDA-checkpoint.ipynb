# Exploratory Data Analysis

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
# Import data
rank_df = pd.read_csv('../data/ranked_data.csv', index_col=0).iloc[:2025]
cat_df = pd.read_csv('../data/cat_data.csv', index_col = 0)
user_df = pd.read_csv('../data/user_ratings.csv')
user_df = user_df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
# Score distribution for ranked data
fig, ax = plt.subplots(1, 2, figsize = (15, 6))

# What is the rating distribution of the top 2000 games?
ax[0].hist(rank_df['avg_rating'], bins = 15)
ax[0].set_title('Rating Distribution for the Top 2000 Games')
ax[0].set_xlabel('Rating from 1 to 10')
ax[0].set_ylabel('Number of Games')
ax[0].axvline(x = rank_df['avg_rating'].mean(), c = 'r',
              label = 'avg_rating = ' + str(round(rank_df['avg_rating'].mean(), 2)))
ax[0].legend()

# What is the relationship between the collection rank and the number of voters?
ax[1].scatter(rank_df['collection_rank'], rank_df['num_voters'])
ax[1].set_title('Collection Rank v. Number of Votes')
ax[1].set_xlabel('Game Ranking')
ax[1].set_ylabel('Number of Votes');
```

```python
# Get ordered dictionary of categories seen in games
categories = Counter(dict(cat_df['cat_1'].value_counts())) +\
             Counter(dict(cat_df['cat_2'].value_counts())) +\
             Counter(dict(cat_df['cat_3'].value_counts())) +\
             Counter(dict(cat_df['cat_4'].value_counts())) +\
             Counter(dict(cat_df['cat_5'].value_counts()))

categories = sorted(categories.items(), key = lambda x:x[1], reverse=True)
categories = categories[0:20]
```

```python
# What categories are represented in the data?
fig, ax = plt.subplots(figsize = (10, 10))

cats = [cat for cat, num in categories]
nums = [num for cat, num in categories]

ax.barh(cats, nums)
ax.invert_yaxis()
ax.set_title('Top 20 Categories Represented in the Top 2000 Games')
ax.set_xlabel('Number of Times Represented')
ax.set_ylabel('Category Name');
```

```python
# Total rating distribution using all user ratings
plt.hist(dict(user_df['rating'].value_counts(normalize = True)), bins = 10);
```

```python
# How does rating distribution change from the highest ranked to the lowest ranked game?
# #1 - 177430, #2025 - 180956
high = user_df[user_df['id'] == 174430]
middle = user_df[user_df['id'] == 8989]
low = user_df[user_df['id'] == 180956]

# Get dictionaries with normalized value counts for scaling
high_vals = dict(high['rating'].value_counts(normalize=True))
middle_vals = dict(middle['rating'].value_counts(normalize=True))
low_vals = dict(low['rating'].value_counts(normalize=True))
```

```python
# Plot overall scores of rating
fig, ax = plt.subplots(figsize = (8, 6))

# Plot rating
plt.hist(high_vals, bins = 10, alpha = 0.4, color = 'black', label = 'Highest Ranked')
plt.hist(middle_vals, bins = 10, alpha = 0.4, color = 'red', label = 'Middle Ranked')
plt.hist(low_vals, bins = 10, alpha = 0.7, color = 'purple', label = 'Lowest Ranked')

# Clean
plt.title('How Rating is Distributed based on Game Ranking')
plt.xlabel('Ranking from 1-10')
plt.ylabel('Rating Distribution')
plt.legend();
```
